```python
import cv2

# Read the noisy image
noisy_image = cv2.imread('random_image.jpg')

# Apply Gaussian blur to smoothen the image
smooth_image = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Save the smooth image
cv2.imwrite('smooth_image.jpg', smooth_image)
print("Smooth image saved successfully.")
```
