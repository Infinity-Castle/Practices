```python
import cv2
import numpy as np

# Upload your image to Google Colab
from google.colab.patches import cv2_imshow

# Read the image
# The second argument 0 specifies that we want to read the image in grayscale mode.
img = cv2.imread('campus.png', 0)  # Read the image in grayscale mode

# Get the negative of the image
neg_img = 255 - img

# Display the original and negative images
cv2_imshow(img)
cv2_imshow(neg_img)
```
