```python
import numpy as np

# Define the OR gate truth table
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 1])

# Initialize weights and bias
w = np.random.randn(2)
b = np.random.randn()

# Define the activation function
def activation(x):
    return 1 if x >= 0 else 0

# Define the Perceptron learning rule
def train(X, y, w, b, learning_rate=0.1, epochs=100):
    for epoch in range(epochs):
        for i in range(len(X)):
            x = X[i]
            y_pred = activation(np.dot(w, x) + b)
            error = y[i] - y_pred
            w += learning_rate * error * x
            b += learning_rate * error
    return w, b

# Train the Perceptron
w, b = train(X, y, w, b)

# Test the Perceptron on random inputs
import random
print("Input\tOutput\tExpected")
for _ in range(10):
    x1 = random.randint(0, 1)
    x2 = random.randint(0, 1)
    x = np.array([x1, x2])
    y_pred = activation(np.dot(w, x) + b)
    expected_output = 1 if x1 == 1 or x2 == 1 else 0
    print(f"{x}\t{y_pred}\t{expected_output}")
```
