```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalize_image(image):
    # Separate the image channels
    b, g, r = cv2.split(image)

    # Convert each channel to floating point representation
    b_float = b.astype(np.float32)
    g_float = g.astype(np.float32)
    r_float = r.astype(np.float32)

    # Normalize each channel independently
    b_normalized = b_float / 255.0
    g_normalized = g_float / 255.0
    r_normalized = r_float / 255.0

    # Merge the normalized channels back into an RGB image
    normalized_image = cv2.merge((r_normalized, g_normalized, b_normalized))

    return normalized_image

# Read the input RGB image
rgb_image = cv2.imread('low.jpeg')

# Normalize each channel of the RGB image
normalized_rgb_image = normalize_image(rgb_image)

# Display the original and normalized images using Matplotlib
plt.figure(figsize=(12, 6))

# Original RGB image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

# Normalized RGB image
plt.subplot(1, 2, 2)
plt.imshow(normalized_rgb_image)
plt.title('Normalized RGB Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalize_image(image):
    # Separate the image channels
    b, g, r = cv2.split(image)

    # Convert each channel to floating point representation
    b_float = b.astype(np.float32)
    g_float = g.astype(np.float32)
    r_float = r.astype(np.float32)

    # Normalize each channel independently
    b_normalized = b_float / 255.0
    g_normalized = g_float / 255.0
    r_normalized = r_float / 255.0

    # Merge the normalized channels back into an RGB image
    normalized_image = cv2.merge((r_normalized, g_normalized, b_normalized))  # Reorder channels to RGB

    return normalized_image

# Read the input RGB image
rgb_image = cv2.imread('low.jpeg')

# Normalize each channel of the RGB image
normalized_rgb_image = normalize_image(rgb_image)

# Display the original and normalized images using Matplotlib
plt.figure(figsize=(12, 6))

# Original RGB image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

# Normalized RGB image
plt.subplot(1, 2, 2)
plt.imshow(normalized_rgb_image)
plt.title('Normalized RGB Image')
plt.axis('off')

plt.show()
```
