```python
import cv2
import matplotlib.pyplot as plt

def apply_threshold(image, threshold_value):
    # Apply binary thresholding
    _, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
    return binary_image

# Read the input grayscale image
gray_image = cv2.imread('gray.jpeg', cv2.IMREAD_GRAYSCALE)

# Define the threshold value (adjust as needed)
threshold_value = 127  # Example threshold value

# Apply thresholding
binary_image = apply_threshold(gray_image, threshold_value)

# Display both images using Matplotlib
plt.figure(figsize=(12, 6))

# Display the grayscale image with label
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Display the binary image with label
plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image after Thresholding')
plt.axis('off')

plt.show()
```
