```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload your image to Google Colab
from google.colab.patches import cv2_imshow

# Read the image
img = cv2.imread('campus.png')  # Read the image in color mode (default)

# Split the image into separate channels
b, g, r = cv2.split(img)

# Compute the histograms for each channel
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

# Plot the histograms
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.plot(hist_b, color='b')
plt.title('Blue Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.plot(hist_g, color='g')
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.plot(hist_r, color='r')
plt.title('Red Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Display the original image
cv2_imshow(img)
```

```python
# hist_b = cv2.calcHist([b], [0], None, [256], [0, 256]), hist_g = cv2.calcHist([g], [0], None, [256], [0, 256]), and hist_r = cv2.calcHist([r], [0], None, [256], [0, 256]): These lines compute the histograms for the individual Blue, Green, and Red channels, respectively, using cv2.calcHist. The [0] parameter specifies that we are computing the histogram for the 0th channel (only one channel). None is a mask, which is not used here. [256] specifies the number of bins (256 for 8-bit images), and [0, 256] specifies the range of pixel values (0 to 255 for 8-bit images).
```
