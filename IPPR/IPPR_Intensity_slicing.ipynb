```python
import cv2
import numpy as np

def intensity_slicing_preserve_background(image, lower_threshold, upper_threshold):
    # Create a mask for the specified intensity range
    mask = cv2.inRange(image, lower_threshold, upper_threshold)

    # Copy the original image
    result = image.copy()

    # Apply the mask to preserve the background
    result[mask != 0] = 255

    return result

def intensity_slicing_non_preserve_background(image, lower_threshold, upper_threshold):
    # Create a mask for the specified intensity range
    mask = cv2.inRange(image, lower_threshold, upper_threshold)

    # Apply the mask without preserving the background
    result = cv2.bitwise_and(image, image, mask=mask)

    return result

# Load the grayscale image
image = cv2.imread('8bit.png', cv2.IMREAD_GRAYSCALE)

# Define the intensity range to highlight
lower_threshold = 100
upper_threshold = 200

# Apply intensity slicing preserving background
result_preserve_background = intensity_slicing_preserve_background(image, lower_threshold, upper_threshold)

# Apply intensity slicing without preserving background
result_non_preserve_background = intensity_slicing_non_preserve_background(image, lower_threshold, upper_threshold)

# Save the results
cv2.imwrite('intensity_slicing_preserve_background.png', result_preserve_background)
cv2.imwrite('intensity_slicing_non_preserve_background.png', result_non_preserve_background)

print("Results saved successfully.")
```
