```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image = cv2.imread('cat.jpeg', cv2.IMREAD_GRAYSCALE)

# Define a convolution kernel (e.g., a simple edge detection kernel)
kernel = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

# Perform convolution
convolved_image = cv2.filter2D(input_image, -1, kernel)

# Define the pooling window size (e.g., 2x2)
pool_size = 2

# Perform max pooling
pooled_image = cv2.resize(convolved_image, (convolved_image.shape[1]//pool_size, convolved_image.shape[0]//pool_size))

# Flatten operation
flattened_features = pooled_image.flatten()

# Print the flattened features
print("Flattened Features Shape:", flattened_features.shape)
print("Flattened Features:", flattened_features)

# Display input and output images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Input Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pooled_image, cmap='gray')
plt.title('Convolved Flattened and Pooled Image')
plt.axis('off')

plt.show()
```
