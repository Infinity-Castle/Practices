```python
import cv2
import numpy as np

# Upload your image to Google Colab
from google.colab.patches import cv2_imshow

# Read the image
img = cv2.imread('campus.png')  # Read the image in color mode (default)

# Get the negative of the image
neg_img = 255 - img

# Display the original and negative images
cv2_imshow(img)
cv2_imshow(neg_img)
```
