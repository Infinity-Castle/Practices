```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform image rotation
def rotate_image(image, angle):
    rows, cols, _ = image.shape

    # Calculate the rotation
    # point of rotation, angle of rotation and scaling factor
    M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)

    # Apply the rotation using warpAffine function
    rotated_img = cv2.warpAffine(image, M, (cols, rows))

    return rotated_img

# Load the image
img_path = "campus.png"  # Change this to the path of your image
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Rotate the image clockwise by 90 degrees
rotated_clockwise_img = rotate_image(image, -90)

# Rotate the image anti-clockwise by 90 degrees
rotated_anticlockwise_img = rotate_image(image, 90)

# Display the original and rotated images
plt.figure(figsize=(20, 15))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(rotated_clockwise_img)
plt.title('Rotated Clockwise (90°)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(rotated_anticlockwise_img)
plt.title('Rotated Anti-clockwise (90°)')
plt.axis('off')

plt.show()
```
