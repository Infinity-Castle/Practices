```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "campus.png"
original_rgb_image = cv2.imread(image_path)
original_rgb_image_float32 = original_rgb_image.astype(np.float32) / 255.0
cmy_image = 1.0 - original_rgb_image_float32
cmy_image_uint8 = (cmy_image * 255).astype(np.uint8)
plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.imshow(cmy_image_uint8)
plt.axis("off")
```
