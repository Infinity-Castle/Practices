```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def average_pooling(input_image, pool_size):
    # Get input image dimensions
    height, width, channels = input_image.shape

    # Calculate output image dimensions after pooling
    output_height = height // pool_size
    output_width = width // pool_size

    # Initialize the output image
    output_image = np.zeros((output_height, output_width, channels), dtype=np.uint8)

    # Perform average pooling
    for y in range(output_height):
        for x in range(output_width):
            for c in range(channels):
                # Calculate the pooling window coordinates
                y_start = y * pool_size
                y_end = y_start + pool_size
                x_start = x * pool_size
                x_end = x_start + pool_size

                # Extract the region from the input image
                region = input_image[y_start:y_end, x_start:x_end, c]

                # Compute the average value of the region
                avg_value = np.mean(region)

                # Assign the average value to the corresponding pixel in the output image
                output_image[y, x, c] = avg_value

    return output_image

# Read the input image
input_image = cv2.imread('image.png')

# Define the pooling window size (e.g., 2x2)
pool_size = 2

# Perform average pooling
output_image = average_pooling(input_image, pool_size)

# Display input and output images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('Output Image')
plt.axis('off')

plt.show()
```

```python
'''
Pooling, also known as subsampling or downsampling, is a technique used in convolutional neural networks (CNNs) and other deep learning architectures for reducing the spatial dimensions (width and height) of feature maps.
Pooling helps in controlling overfitting, reducing computational complexity, and improving the computational efficiency of the model.

There are different types of pooling operations, such as max pooling, average pooling, and min pooling. Among them, max pooling and average pooling are the most commonly used.

Max Pooling: In max pooling, for each region in the input feature map, the maximum value is selected and retained while the rest are discarded.
This helps in retaining the most important features present in the region.

Average Pooling: In average pooling, for each region in the input feature map, the average value of all the elements in the region is computed and retained.
This helps in smoothing out the features and reducing noise.

Pooling is typically applied after convolutional layers in CNNs. It helps in reducing the spatial dimensions of the feature maps, leading to a decrease in the number of parameters and computational complexity in subsequent layers.
Additionally, pooling helps in creating feature maps that are more robust to spatial translations and variations in the input data.

'''
```
