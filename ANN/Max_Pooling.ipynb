```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def max_pooling(input_image, pool_size):
    # Get input image dimensions
    height, width, channels = input_image.shape

    # Calculate output image dimensions after pooling
    output_height = height // pool_size
    output_width = width // pool_size

    # Initialize the output image
    output_image = np.zeros((output_height, output_width, channels), dtype=np.uint8)

    # Perform max pooling
    for y in range(output_height):
        for x in range(output_width):
            for c in range(channels):
                # Calculate the pooling window coordinates
                y_start = y * pool_size
                y_end = y_start + pool_size
                x_start = x * pool_size
                x_end = x_start + pool_size

                # Extract the region from the input image
                region = input_image[y_start:y_end, x_start:x_end, c]

                # Compute the maximum value of the region
                max_value = np.max(region)

                # Assign the maximum value to the corresponding pixel in the output image
                output_image[y, x, c] = max_value

    return output_image

# Read the input image
input_image = cv2.imread('cat.jpeg')

# Define the pooling window size (e.g., 2x2)
pool_size = 2

# Perform max pooling
output_image = max_pooling(input_image, pool_size)

# Display input and output images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title('Input Image ({}x{})'.format(input_image.shape[1], input_image.shape[0]))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('Output Image ({}x{})'.format(output_image.shape[1], output_image.shape[0]))
plt.axis('off')

plt.show()
```
