```python
import numpy as np
import time

def threshold(x):
    if x >= 0:
        return 1
    else:
        return -1

def print_func(loop_var, net, sig_net, teacher_signal, w, delta_w=None):
    print("i:", loop_var)
    print("net:", net)
    print("sig_net:", sig_net)
    print("delta_w:", delta_w)
    print("teacher_signal:", teacher_signal)
    print("w:", w)
    print("-------------------\n")

def compute():
    try:
        n = int(input("Enter number of input vectors: "))
        x = []
        c = 0.1
        for i in range(0, n):
            raw_str1 = input("Enter values for vector {}: ".format(i+1))
            input_vector = raw_str1.split(' ')
            ip_list = [float(ele) for ele in input_vector]
            x.append(np.array(ip_list, dtype=np.float64))

        raw_str2 = input("Enter values for teacher signal: ")
        teacher_signal = [int(x) for x in raw_str2.split(' ')]
        if len(teacher_signal) != n:
            print("Teacher Signal length Error..")
            time.sleep(3)
            exit()

        raw_str3 = input("Enter initial weight vector: ")
        w_list = [float(ele) for ele in raw_str3.split(' ')]

        for i in range(0, n):
            net = np.dot(w_list, x[i])
            sig_net = threshold(net)

            if sig_net != teacher_signal[i]:
                delta_w = c * (teacher_signal[i] - sig_net) * x[i]
                w_list = np.add(w_list, delta_w)

            print_func(i, net, sig_net, teacher_signal[i], w_list, delta_w)

    except Exception as e:
        print("Error:", str(e))

if __name__ == '__main__':
    compute()
```
