```python
import cv2
import numpy as np

# Load the two grayscale images
image1 = cv2.imread('figure.jpg', cv2.IMREAD_GRAYSCALE)
image2 = cv2.imread('figure.jpg', cv2.IMREAD_GRAYSCALE)

# Ensure images are of the same size
if image1.shape != image2.shape:
    print("Images must be of the same size.")
    exit()

# Addition of two images
addition_result = cv2.add(image1, image2)

# Subtraction of two images
subtraction_result = cv2.subtract(image1, image2)

# Save the results
cv2.imwrite('figure.png', addition_result)
cv2.imwrite('subtraction_result.png', subtraction_result)

print("Results saved successfully.")
```
