```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to perform image translation
def translate_image(image, x, y):
    rows, cols, channels = image.shape

    # Define the translation matrix
    #The first two elements of each row ([1, 0] and [0, 1]) represent the scaling factors for the x and y coordinates, respectively. Since they are set to 1, there is no scaling applied.
    M = np.float32([[1, 0, x], [0, 1, y]])

    # Apply the translation using warpAffine function
    translated_img = cv2.warpAffine(image, M, (cols, rows))

    return translated_img

# Load the image
img_path = "campus.png"  # Change this to the path of your image
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Shift the image to the right by 20 units
translated_right_img = translate_image(image, 100, 0)

# Shift the image downwards by 10 units
translated_down_img = translate_image(image, 0, 100)

# Display the original and translated images
plt.figure(figsize=(15, 15))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(translated_right_img)
plt.title('Shifted Right')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(translated_down_img)
plt.title('Shifted Downwards')
plt.axis('off')
plt.show()
```
