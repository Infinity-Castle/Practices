```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

def plot_histogram(img, title):
    # Split the image into channels
    b, g, r = cv2.split(img)

    # Equalize each channel
    equalized_b = cv2.equalizeHist(b)
    equalized_g = cv2.equalizeHist(g)
    equalized_r = cv2.equalizeHist(r)

    # Merge the equalized channels
    equalized_img = cv2.merge((equalized_b, equalized_g, equalized_r))

    # Calculate histograms
    hist_original_b = cv2.calcHist([b], [0], None, [256], [0, 256])
    hist_original_g = cv2.calcHist([g], [0], None, [256], [0, 256])
    hist_original_r = cv2.calcHist([r], [0], None, [256], [0, 256])

    hist_equalized_b = cv2.calcHist([equalized_b], [0], None, [256], [0, 256])
    hist_equalized_g = cv2.calcHist([equalized_g], [0], None, [256], [0, 256])
    hist_equalized_r = cv2.calcHist([equalized_r], [0], None, [256], [0, 256])

    # Plot histograms
    plt.figure(figsize=(15, 10))
    plt.subplot(3, 4, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(3, 4, 2)
    plt.plot(hist_original_b, color='b')
    plt.plot(hist_original_g, color='g')
    plt.plot(hist_original_r, color='r')
    plt.title('Original Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    plt.subplot(3, 4, 5)
    plt.imshow(cv2.cvtColor(equalized_img, cv2.COLOR_BGR2RGB))
    plt.title('Equalized Image')
    plt.axis('off')

    plt.subplot(3, 4, 6)
    plt.plot(hist_equalized_b, color='b')
    plt.plot(hist_equalized_g, color='g')
    plt.plot(hist_equalized_r, color='r')
    plt.title('Equalized Histogram')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    plt.suptitle(title)
    plt.show()

# Load the image
img_path = "city.jpeg"  # Change this to the path of your image
img = cv2.imread(img_path)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot original and equalized histograms
plot_histogram(img, 'Original and Equalized Histograms')
```
