```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convolution(input_image, kernel):
    # Perform convolution
    output_image = cv2.filter2D(input_image, -1, kernel)
    return output_image

def max_pooling(input_image, pool_size):
    # Get input image dimensions
    height, width = input_image.shape

    # Calculate output image dimensions after pooling
    output_height = height // pool_size
    output_width = width // pool_size

    # Initialize the output image
    output_image = np.zeros((output_height, output_width), dtype=np.uint8)

    # Perform max pooling
    for y in range(output_height):
        for x in range(output_width):
            # Calculate the pooling window coordinates
            y_start = y * pool_size
            y_end = y_start + pool_size
            x_start = x * pool_size
            x_end = x_start + pool_size

            # Extract the region from the input image
            region = input_image[y_start:y_end, x_start:x_end]

            # Compute the maximum value of the region
            max_value = np.max(region)

            # Assign the maximum value to the corresponding pixel in the output image
            output_image[y, x] = max_value

    return output_image

# Read the input image
input_image = cv2.imread('cat.jpeg')

# Convert the input image to grayscale
input_image_gray = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)

# Define a kernel for convolution (e.g., a simple edge detection kernel)
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])

# Perform convolution
convolved_image = convolution(input_image_gray, kernel)

# Define the pooling window size (e.g., 2x2)
pool_size = 2

# Perform max pooling
pooled_image = max_pooling(convolved_image, pool_size)

# Display input and output images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(convolved_image, cmap='gray')
plt.title('Convolved Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(pooled_image, cmap='gray')
plt.title('Pooled Image')
plt.axis('off')

plt.show()
```
