```python
import cv2
from google.colab.patches import cv2_imshow

def scale_image(image, scale_factor):
  # Handle potential conversion from BGR to RGB if needed
  if cv2.cvtColor(image, cv2.COLOR_BGR2RGB).shape != image.shape:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

  # Get the original image dimensions
  height, width, channels = image.shape

  # Calculate the new dimensions based on the scaling factor
  new_height = int(height * scale_factor)
  new_width = int(width * scale_factor)

  # Use interpolation for smoother scaling
  resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_AREA)

  return resized_image

# Load the image (replace 'path/to/your/image.jpg' with your image path)
image = cv2.imread('campus.png')

# Scale the image by a factor of 2
scaled_image_2x = scale_image(image.copy(), 2.0)

# Scale the image by a factor of 0.5
scaled_image_05x = scale_image(image.copy(), 0.5)

# Display the original and scaled images
cv2_imshow(scaled_image_05x)
cv2_imshow(image)
cv2_imshow(scaled_image_2x)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
